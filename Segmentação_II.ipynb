Carregar a imagem 'chips.png'.

````python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image = mpimg.imread("chips.png")
flat_image=(image[:,:,0]+image[:,:,1]+image[:,:,2])/3.

r, c = np.shape(flat_image)
gd = 4
test_slice = flat_image[::gd,::gd]
fig,ax=plt.subplots(1,1)
the_image = ax.imshow(
                flat_image,
                zorder=0,alpha=1.0,
                cmap="Greys_r",
                origin="upper",
                interpolation="hermite",
            )
plt.colorbar(the_image)            
Y, X = np.mgrid[0:r:gd, 0:c:gd]
dY, dX = np.gradient(test_slice)
ax.quiver(X, Y, dX, dY, color='r')
plt.show()
````

Calcular o gradiente e mostrar uma imagem com os vetores sobrepostos à imagem 'chips.png', utilizando a função 'quiver'.
Obter a magnitude do gradiente e limiarizá-la.

````python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('chips.png',0)
image = cv2.medianBlur(image,5)

ret,th1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [image, th1, th2, th3]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
````

````python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('chips.png',0)
edges = cv2.Canny(image,100,200)
plt.subplot(121),plt.imshow(image,cmap = 'gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Imagem modificada'), plt.xticks([]), plt.yticks([])

plt.show()
````

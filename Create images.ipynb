```python
%matplotlib inline
import numpy as ny
import matplotlib.pyplot as pl
from scipy import signal
```

1) Create the image of a paraboloid with one axis scaled (like an oval paraboloid).

```python
def paraboloid(x, y, a, b, c): 
    return c*((pow(x,2)/pow(a,2)) - (pow(y,2)/pow(b,2)))

x = ny.linspace(-10, 10, 300)
y = ny.linspace(-10, 10, 300)

X, Y = ny.meshgrid(x, y)
Z = paraboloid(X, Y, 1, 1, 1)

fig = pl.figure()
ax = pl.axes(projection='3d')
ax.contour3D(X, Y, Z, 80, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

2) Create the image of a rotated sin(x) using rotation of coordinates.

```python
#Rotation one
fig = pl.figure()
ax = pl.axes(projection='3d')

x = ny.linspace(-2*ny.pi, 2*ny.pi, 200)
y = ny.linspace(-2*ny.pi, 2*ny.pi, 200)
tau, phase = ny.meshgrid(x, y) 
amp = ny.sin(tau + phase)

ax.contour3D(x, y, amp, 80, cmap='binary')
ax.set_xlabel('tau')
ax.set_ylabel('phase')
ax.set_zlabel('ampli')

#rotation two
fig = pl.figure()
ax = pl.axes(projection='3d')
 
tau_rot = ny.cos(0.50)*tau + ny.sin(0.50)*phase 
phase_rot = -ny.sin(0.50)*tau + ny.cos(0.50)*phase

ax.contour3D(tau_rot, phase_rot, amp, 80, cmap='binary')

ax.set_xlabel('tau')
ax.set_ylabel('phase')
ax.set_zlabel('ampli')
```

3) Create the image of a gaussian in 2D.

```python

N = 9 
k1d = signal.gaussian(N, std=1).reshape(N, 1)
kernel = ny.outer(k1d, k1d)
pl.imshow(kernel)
pl.show()

A = ny.zeros((16, 16))
A[5, 9] = 1
pl.imshow(A)
pl.show()

row, col = ny.where(A == 1)
A[row[0]-(N//2):row[0]+(N//2)+1, col[0]-(N//2):col[0]+(N//2)+1] = kernel
pl.imshow(A)
```

```python
from google.colab import drive
from google.colab.patches import cv2_imshow
from PIL import Image
import numpy as ny
import cv2
```

# ***(1) Crop and flip an image using Numpy array indexing.***

```python
espada = ny.array(Image.open('excalibur.jpg'))
print(espada.shape)
img_crop = espada[200:400, 200:400]
Image.fromarray(img_crop).save('excalibur.jpg')
Image.fromarray(ny.flipud(espada)).save('excalibur.jpg')

img_crop = cv2.cvtColor(img_crop, cv2.COLOR_BGR2RGB)
img_flipud = cv2.cvtColor(ny.flipud(espada), cv2.COLOR_BGR2RGB)
cv2_imshow(img_crop)
cv2_imshow(img_flipud)
```

# ***(2) Implement image translation using Numpy and OpenCV.***

```python
espada1 = cv2.imread('excalibur.jpg')

largura = espada1.shape[1]
altura = espada1.shape[1]
matx = ny.float32([[1, 0, 100], [0, 1, 200]])
img_translation = cv2.warpAffine(espada1, matx, (largura, altura))

cv2.imwrite('excalibur.jpg', img_translation)
cv2_imshow(img_translation)
```

# ***(3) Implement image rotation using Numpy and OpenCV.***

```python
espadaRotação = cv2.rotate(espada1, cv2.ROTATE_90_CLOCKWISE)
cv2.imwrite('excalibur.jpg', espadaRotação)

cv2_imshow(espadaRotação)
Image.fromarray(ny.rot90(espada, 2)).save('excalibur.jpg')
img_180 = cv2.cvtColor(ny.rot90(espada, 2), cv2.COLOR_BGR2RGB)
cv2_imshow(img_180)
```

# ***(4) Implement image resizing using OpenCV.***

```python
siz_largura = int(largura * 0.2)
siz_altura = int(altura * 0.2)

dsize = (siz_largura, siz_altura)
img_siz = cv2.resize(espada1, dsize)
print(img_siz.shape)

cv2.imwrite('excalibur1.jpg', img_siz)
cv2_imshow(img_siz)
```

# ***(5) Implement bitwise operations: AND, OR, XOR.***

```python
rectangle = ny.zeros((300, 300), dtype="uint8")
cv2.rectangle(rectangle, (25, 25), (275, 275), 255, -1)

# draw a circle
circle = ny.zeros((300, 300), dtype = "uint8")
cv2.circle(circle, (150, 150), 150, 255, -1)

bitw_and = cv2.bitwise_and(rectangle, circle)
bitw_or = cv2.bitwise_or(rectangle, circle)
bitw_xor = cv2.bitwise_xor(rectangle, circle)

cv2.imwrite('/content/drive/MyDrive/images/and.jpg', bitw_and)
cv2.imwrite('/content/drive/MyDrive/images/or.jpg', bitw_or)
cv2.imwrite('/content/drive/MyDrive/images/xor.jpg', bitw_xor)

cv2_imshow(bitw_and)
cv2_imshow(bitw_or)
cv2_imshow(bitw_xor)
```

# ***(6) Implement the "mask" operation, where a third image 'h' contains only a Region of Interest (ROI -- defined by the second image mask 'g') obtained from the input image 'f'. Note that this Region can be of any shape.***

```python
espadaMask = cv2.imread('excalibur.jpg')
circle = ny.zeros(espadaMask.shape[:2], dtype = ny.uint8)
cv2.circle(circle, (150, 150), 150, 255, -1)

espadaRealMask = cv2.bitwise_and(espadaMask, espadaMask, mask=circle)

cv2.imwrite('cicurlo.jpg', espadaRealMask)
cv2_imshow(espadaRealMask)
```
